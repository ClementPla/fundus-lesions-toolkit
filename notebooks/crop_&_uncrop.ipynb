```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import cv2
sys.path.append('../lib/fundus_lesions_toolkit')
from utils.images import open_image, plot_image, autocrop_fundus, pad_given_margins, autofit_fundus_resolution, reverse_autofit
```

```python
example_path = '../imgs/rfmid_sample_1.png'
img = open_image(example_path)
plot_image(img, title='Sample from RFMID')
cropped_img, roi, margins = autocrop_fundus(img)
plot_image(cropped_img, title='Cropped Image')

padded_img = pad_given_margins(cropped_img, margins)
plot_image(padded_img, title='Reverse cropping')

autofit_fundus, params = autofit_fundus_resolution(img, 512)
plot_image(autofit_fundus, title='Autofit')

back_to_original_fundus = reverse_autofit(autofit_fundus, **params)
plot_image(back_to_original_fundus, title='Back to original')
```
