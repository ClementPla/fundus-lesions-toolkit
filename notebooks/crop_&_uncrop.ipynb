# Simple cropping tools dedicated to fundus

```python
%load_ext autoreload
%autoreload 2
```

```python
from fundus_lesions_toolkit.utils.images import open_image, autocrop_fundus, pad_given_margins, autofit_fundus_resolution, reverse_autofit
from fundus_lesions_toolkit.utils.visualization import plot_image
```

```python
example_path = '../imgs/rfmid_sample_1.png'
img = open_image(example_path)
plot_image(img, title=f'Sample from RFMID {img.shape}', figsize=(5,5))

print("Autocrop fundus (removal of black borders)")
cropped_img, roi, margins = autocrop_fundus(img)
plot_image(cropped_img, title=f'Cropped Image {cropped_img.shape}', figsize=(5,5))

print("Reverse autocrop (pad with black borders)")
padded_img = pad_given_margins(cropped_img, margins)
plot_image(padded_img, title=f'Reverse cropping {padded_img.shape}', figsize=(5,5))

print("Autofit: Autocrop, resize and pad to square resolution (512x512)")
autofit_fundus, params = autofit_fundus_resolution(img, 512)
plot_image(autofit_fundus, title=f'Autofit {autofit_fundus.shape}', figsize=(5,5))

print("Reverse autofit")
back_to_original_fundus = reverse_autofit(autofit_fundus, **params)
plot_image(back_to_original_fundus, title=f'Back to original {back_to_original_fundus.shape}', figsize=(5,5))
```
